# LR using Gradient Decent Optimizer

```python
m = 0
c = 0
l = 0.0001

iterations = 1000

n = int(len(x))
          
for i in range(iterations):
          y_pred =  (m*x) + c
          derivatives_m = (-2/n) * sum(x*(y-y_pred))
          derivatives_c = (-2/n) * sum(y-y_pred)
          
          m = m-(l*derivatives_m)
          c = c-(l*derivatives_c)
          
          print(c,m)
```
