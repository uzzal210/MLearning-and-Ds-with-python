# One Hot Encoding

# Doing all these using for loop

```python
df=pd.read_csv('Pizza Data.csv')
df.head()
```

```python
columns=['Company']
for col in columns:
    one=pd.get_dummies(df[col])##creating dummy variable
    df=pd.concat((df,one),axis=1).drop(col,axis=1)
```

```python
df.head()
```

# Ordinal Encoder

```python
df=pd.read_csv('Pizza Data.csv')
df.head()
```

```python
name=["Domino's Pizza", 'Pizza Hut', "Godfather's Pizza", "IMO's Pizza"]
```

```python
ordinal=OrdinalEncoder(categories=[name])
encoded = ordinal.fit_transform(df[['Company']])
```
