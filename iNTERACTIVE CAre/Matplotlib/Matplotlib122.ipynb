# Line plot

```python
import numpy as np
from matplotlib  import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
```

# Adding titles and labels

```python
plt.plot(x,y)
plt.title('Line plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=5)
```

```python
plt.plot(x,y,'g',linestyle=':',linewidth=5,label='SLine')
plt.legend()
```

# Adding two  lines in the same plot:

```python
y1=2*x
y1
```

```python
y2=3*x
y2
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=2,label='one line')
plt.plot(x,y2,color='r',linestyle='-',linewidth=2,label='two line')
plt.title('Line plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True,color='k')
#plt.grid(True)
```

# Adding subplot

```python
x
#y1
#y2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle=':',linewidth=2)

```

# Bar plot

```python
student={'Bob':87,'Matt':56,'Sam':27}
names=list(student.keys())
values=list(student.values())
plt.bar(names,values)
```

# Adding title and labels

```python
plt.bar(names,values,color='m')
plt.xlabel('Name')
plt.ylabel('Score')
```

```python
plt.barh(names,values,color='g')
plt.xlabel('Score')
plt.ylabel('Name')
```

# Scatter plot

```python
plt.figure(figsize=(10,10))
x=[1,2,3,4,5,6,7,8,9]
a=[8,1,7,2,0,3,7,3,2]
plt.scatter(x,a)
```

# Changing Mark Aesthetics

```python
plt.figure(figsize=(10,10))
x=[1,2,3,4,5,6,7,8,9]
a=[8,1,7,2,0,3,7,3,2]
plt.scatter(x,a,marker='*',c='g',s=100)
```

# Adding two markers in the same plot

```python
plt.figure(figsize=(10,10))
x=[1,2,3,4,5,6,7,8,9]
a=[8,1,7,2,0,3,7,3,2]
b=[7,2,5,6,9,1,4,5,3]
plt.scatter(x,a,marker='*',c='g',s=100,label='one')
plt.scatter(x,b,marker='^',c='r',s=100,label='two')
plt.legend()
```

# Adding subplot

```python
plt.figure(figsize=(10,10))
x=[1,2,3,4,5,6,7,8,9]
a=[8,1,7,2,0,3,7,3,2]
b=[7,2,5,6,9,1,4,5,3]
plt.scatter(1,2,1)
plt.scatter(x,a,marker='*',c='g',s=100,label='one')
plt.scatter(1,2,2)
plt.scatter(x,b,marker='^',c='r',s=100,label='two')
plt.legend()
```

# Box plot

```python
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]
data=list([one,two,three])
plt.boxplot(data)
```

# Violin plot

```python
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]
data=list([one,two,three])
plt.violinplot(data,showmedians=True)
#plt.violinplot(data)
```

# pie chart

```python
fruit=['apple','orange','mango','guava']
quantity=[67,34,100,29]
plt.pie(quantity,labels=fruit)
```

# change aesthetics

```python
fruit=['apple','orange','mango','guava']
quantity=[67,34,100,29]
plt.pie(quantity,labels=fruit,autopct='%.4f%%',colors=['r','g','b','m'],startangle=90,shadow=True,explode=(.1,0.1,0.1,0.1,))
```

# DoughNut Chart

```python
fruit=['apple','orange','mango','guava']
quantity=[67,34,100,29]
plt.pie(quantity,labels=fruit,radius=2)
plt.pie([1],colors=['k'],radius=1)
```
